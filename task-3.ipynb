```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/job-description-dataset/job_descriptions.csv")
```

```python
def extract_min_exp(exp_str):
    try:
        parts = exp_str.split("to")
        min_exp = int(parts[0].strip().split()[0])
        return min_exp
    except:
        return None

```

```python
df["min_exp"] = df["Experience"].apply(extract_min_exp)

```

```python
def parse_min_salary(sal):
    try:
        sal = sal.replace("$", "").replace("K", "").replace(",", "")
        min_sal = int(sal.split("-")[0])
        return min_sal
    except:
        return None
```

```python
df["min_salary"] = df["Salary Range"].apply(parse_min_salary)
df = df[df["min_salary"] > 50]

```

```python
asian_countries = [
    'India', 'China', 'Japan', 'Korea, Rep.', 'Singapore', 'Malaysia', 'Thailand', 'Indonesia',
    'Pakistan', 'Bangladesh', 'Vietnam', 'Sri Lanka', 'Nepal', 'Philippines',
    'UAE', 'Qatar', 'Saudi Arabia', 'Afghanistan', 'Iran, Islamic Rep.',
    'Iraq', 'Kazakhstan', 'Bhutan', 'Myanmar', 'Uzbekistan', 'Maldives',
    'Mongolia', 'Armenia', 'Azerbaijan', 'Bahrain', 'Brunei', 'Cambodia',
    'Georgia', 'Kyrgyz Republic', 'Lao PDR', 'Lebanon', 'Syrian Arab Republic',
    'Tajikistan', 'Timor-Leste', 'Turkmenistan', 'Turkey', 'Yemen'
]
```

```python
filtered_df = df[
    (df['Company Size'] < 50000) &
    (df['Job Title'].str.lower().str.strip() == 'Mechanical Engineer') &
    (df['min_exp'] >= 5) &
    (df['Country'].isin(asian_countries)) &
    (df['min_salary'] > 50000) &
    (df['Work Type'].str.lower().isin(['Full-Time', 'Part-Time'])) &
    (df['Preference'].str.lower().str.strip() == 'Male') &
    (df['Job Portal'].str.lower().str.strip() == 'Idealist')
]
```
