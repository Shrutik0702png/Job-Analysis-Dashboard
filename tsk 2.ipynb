```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/job-description-dataset/job_descriptions.csv')
```

```python
df['Role'] = df['Role'].astype(str).str.strip()
df['Job Title'] = df['Job Title'].astype(str).str.strip()
df['Preference'] = df['Preference'].astype(str).str.strip()
df['Country'] = df['Country'].astype(str).str.strip()
df['Qualifications'] = df['Qualifications'].astype(str).str.strip()
```

```python
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'], format='%Y-%m-%d', errors='coerce')
```

```python
asian_countries = ['Afghanistan', 'Armenia', 'Azerbaijan', 'Bahrain', 'Bangladesh', 'Bhutan', 'Brunei', 'Cambodia', 'China', 'Georgia', 'Hong Kong SAR, China', 'India', 'Indonesia', 'Iran, Islamic Rep.', 'Iraq', 'Israel', 'Japan', 'Jordan', 'Kazakhstan', 'Korea, Rep.', 'Kuwait', 'Kyrgyz Republic', 'Lao PDR', 'Lebanon', 'Malaysia', 'Maldives', 'Mongolia', 'Myanmar', 'Nepal', 'North Korea', 'Oman', 'Pakistan', 'Philippines', 'Qatar', 'Saudi Arabia', 'Singapore', 'Sri Lanka', 'Syrian Arab Republic', 'Tajikistan', 'Thailand', 'Timor-Leste', 'Turkey', 'Turkmenistan', 'UAE', 'Uzbekistan', 'Vietnam', 'Yemen', 'West Bank and Gaza']
```

```python
filtered_df = df[
    (df['Role'] == 'Data Engineer') &
    (df['Job Title'] == 'Data Scientist') &
    (df['Preference'].str.lower() == 'female') &
    (~df['Country'].isin(asian_countries)) &
    (~df['Country'].str.startswith('C')) &
    (df['latitude'] >= 10) &
    (df['Qualifications'] == 'B.Tech') &
    (df['Job Posting Date'] >= '2023-01-01') &
    (df['Job Posting Date'] <= '2023-06-01')
]
```

```python
print("Filtered rows:", filtered_df.shape[0])
filtered_df.head(10)
```

```python
filtered_df.to_csv('/kaggle/working/task2_filtered_data.csv', index=False)
```

```python
top_10_companies = (
    filtered_df['Company']
    .value_counts()
    .head(10)
    .index
)

final_df = filtered_df[filtered_df['Company'].isin(top_10_companies)]

final_df.to_csv('/kaggle/working/task2_filtered_top10.csv', index=False)
```
