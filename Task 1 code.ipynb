```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/job-description-dataset/job_descriptions.csv')
```

```python
filtered_df = df[
     (df['Role'] == 'User Experience Designer') &
     (df['Job Title'] == 'UX/UI Designer') &
     (df['Company'].str.len() > 5)
]
```

```python
company_counts = (
    filtered_df['Company']
    .value_counts()
    .head(20)
)
```

```python
top_20_unique = filtered_df[filtered_df['Company'].isin(company_counts.index)] \
.drop_duplicates(subset='Company')\
.head(20)
```

```python
top_20_unique.to_csv('top_20_companies.csv', index=False)
```
